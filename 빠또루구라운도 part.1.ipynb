결측치 하나 존재하므로 제거함.

```python
train.dropna(axis=0)
```

```python
#플레이어 순위 산정에 필요없는 column 제거
train.drop(["Id", "matchId", "matchType", "rankPoints", "killPoints", "winPoints"], axis=1, inplace=True)
```

```python
train
```

```python
#그룹별로 묶어 평균을 구함.
train = train.groupby(['groupId'], as_index = False).mean().round(3)
```

```python
#분석에 필요없는 변수이므로 제외
train.drop(["groupId"], axis=1, inplace=True)
train
```
